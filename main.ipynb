```python
# 分片
from typing import List
# 創建函數 處理分片邏輯 讀取文件->content變數
def split_into_chunks(doc_file: str) -> List[str]:
    with open(doc_file, 'r',encoding='utf-8') as file:
        content = file.read()  
    #按行分割
    return [chunk for chunk in content.split("\n\n")]  
    
chunks = split_into_chunks("doc.md")

#迴圈印出片段跟對應編號
for i, chunk in enumerate(chunks):
    print(f"[{i}] {chunk}\n")
```

```python
#索引
from sentence_transformers import SentenceTransformer
#加載emberdding model
embedding_model = SentenceTransformer("shibing624/text2vec-base-chinese")
#獲取某個給定片段的向量 並且將它返回
def embed_chunk(chunk: str) -> List[float]:
    embedding = embedding_model.encode(chunk, normalize_embeddings=True)
    return embedding.tolist()
#測試
embedding = embed_chunk("測試")
print(len(embedding))
print(embedding)
```

```python
#調用函數 給所有片段都生成對應向量
embeddings = [embed_chunk(chunk) for chunk in chunks]

#印出長度跟第一個片段向量
print(len(embeddings))
print(embeddings[0])
```

```python
#將向量存入向量資料庫
import chromadb
#選用內存型向量資料庫 結束後會清除
chromadb_client = chromadb.EphemeralClient()
# chromadb.PersistentClient("./chroma.db") 存入硬碟
#創建表格
chromadb_collection = chromadb_client.get_or_create_collection(name="default")
#把所有片段內容、片段向量、片段ID 一起->chromadb
def save_embeddings(chunks: List[str], embeddings: List[List[float]]) -> None:
    for i, (chunk, embedding) in enumerate(zip(chunks, embeddings)):
        chromadb_collection.add(
            documents=[chunk],
            embeddings=[embedding],
            ids=[str(i)]
        )
#調用函數
save_embeddings(chunks, embeddings)
```

```python
#召回
#query 使用者問題 top_k召回的紀錄數量
def retrieve(query: str, top_k: int) -> List[str]:
    query_embedding = embed_chunk(query) #把使用者問題轉向量
    #把向量傳入chromadb裡面 查詢與使用者問題最像似的前top_k結果
    results = chromadb_collection.query(
        query_embeddings=[query_embedding],
        n_results=top_k
    )
    return results['documents'][0] #把相似的片段撈出來

query = "小茹是誰？"
retrieved_chunks = retrieve(query, 5) #找出5個相似片段
for i, chunk in enumerate(retrieved_chunks):
    print(f"[{i}] {chunk}\n")
```

```python
#重排 
from sentence_transformers import CrossEncoder
#query用戶問題  retrieved_chunks召回的片段 top_k重排後保留的數量
def rerank(query: str, retrieved_chunks: List[str], top_k: int) -> List[str]:
    cross_encoder = CrossEncoder('cross-encoder/mmarco-mMiniLMv2-L12-H384-v1') #精確排序
    pairs = [(query, chunk) for chunk in retrieved_chunks] #用戶問題+召回片段
    scores = cross_encoder.predict(pairs) #發回模型給片段打分

    scored_chunks = list(zip(retrieved_chunks, scores)) #把召回片段跟分數放一起 並產生列表
    scored_chunks.sort(key=lambda x: x[1], reverse=True) #reverse=True分數高的放在第一名

    return [chunk for chunk, _ in scored_chunks][:top_k] #再取top_k個結果片段

reranked_chunks = rerank(query, retrieved_chunks, 3) #只取前三個

for i, chunk in enumerate(reranked_chunks):
    print(f"[{i}] {chunk}\n")
```

```python
#LLM
from dotenv import load_dotenv
from google import genai

load_dotenv()
google_client = genai.Client()
#使用者問題 quert 重排後的片段 chunks
def generate(query: str, chunks: List[str]) -> str:
    prompt = f"""你是一位知識助理，請根據使用者問題和下列片段來生成準確地回答。
使用者問題: {query}

相關片段:
{"\n\n".join(chunks)}

請基於上述內容做答，不要編造訊息"""
    print(f"{prompt}\n\n---\n")
    #調用gemini-2.5-flash模型
    response = google_client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )

    return response.text

answer = generate(query, reranked_chunks)
print(answer)
```
